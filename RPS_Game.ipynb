```python
from tkinter import *
from random import randint 
from tkinter import ttk

root = Tk()
root.title('Rock, Paper, Scissors ')
root.geometry("500x600")
bg = PhotoImage( file = "img.png")
label1 = Label( root, image = bg)
label1.place(x = 350,y = 80)
frame1 = Frame(root, bg = "#88cffa")
frame1.pack(pady = 20)



#pick random no between 0 and 2
pick_number = randint(0,2)

#create comp_choice fun
def spin():
    pick_number = randint(0,2)
    label.config(text = r_choice[pick_number])
    
    if user_choice.get() == "Rock":
        user_value = 0
    if user_choice.get() == "Paper":
        user_value = 1
    if user_choice.get() == "Scissors":
        user_value = 2
        
    
    if(user_value == 0):
        if(pick_number == 0):
            win_label.config(text="It's a Tie")
            
        elif(pick_number == 1):
            win_label.config(text="Computer Wins")
            
        else:
            win_label.config(text="You Win")
            
    if(user_value == 1):
        if(pick_number == 0):
            win_label.config(text="You Win")
            
        elif(pick_number == 1):
            win_label.config(text="It's a Tie")
           
        else:
            win_label.config(text="Computer wins")
            
    if(user_value == 2):
        if(pick_number == 0):
            win_label.config(text="Computer Wins")
            
        elif(pick_number == 1):
            win_label.config(text="You win")
            
        else:
            win_label.config(text="It's a Tie")


    

#List of choices
r_choice = ['Rock','Paper','Scissor']


#user choice
l=Label(root,text="Enter Your Choice")
l.pack(pady=20)
user_choice = ttk.Combobox(root, value=("Rock","Paper","Scissors"))
user_choice.pack(pady=20)

# choice button for comp
choice_button = Button(root, text="Computer Turn", command=spin)
choice_button.pack(pady=10)


#throw output when program runs
label = Label(root, text = r_choice[pick_number])
label.pack(pady=20)


#win or not label
win_label = Label(root,text="",font=("Helvetica",18))
win_label.pack(pady=50)


root.mainloop()
```
